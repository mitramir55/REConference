```python
import json
import pandas as pd
import requests
import pprint

INPUT = "./Data/TasksAndEnhancements.json"
REST_API = "https://bugzilla.mozilla.org/rest/bug/" # 1575284
STAGE1_OUTPUT = "./data/FirefoxDataset.csv"


```

```python
new_url = REST_API+"1575284"
print ("Request call", new_url)
resp = requests.get(new_url)
ele = resp.json()

print(ele['bugs'][0]['blocks'])
pprint.pprint(ele)
while (1):
    stringIs = input()
    
    print(stringIs,": ",ele['bugs'][0][stringIs])
```

```python
def ReadInput():
    with open(INPUT, 'r', encoding='utf8') as f:
        distros_dict = json.load(f)
#     pprint.pprint(distros_dict)
    #input("hit enter to proceed")
    list_elmts = distros_dict['bugs']
    df_Data = pd.DataFrame(list_elmts)
    
    return df_Data


def FetItemInfo(id):
    new_url = REST_API+str(id)
    try:
        resp = requests.get(new_url)
        ele = resp.json()
        return ele['bugs'][0]
    
    except Exception:
        return 0
    
def fetchAllrecs(eleList):
    BigList = []
    count = 1
    for i in eleList:
        count = count +1
        inst = FetItemInfo(i)
        if inst!=0:
            BigList.append(FetItemInfo(i))
            if count == 50:
                break
            print (count)
    return BigList
    
```

```python
def main():
    df_data = ReadInput()
    #print(df_data.head)
    AllData = fetchAllrecs(df_data.id)    
    df_AllData = pd.DataFrame(AllData)
    print(df_AllData.head)
    df_AllData.to_csv(STAGE1_OUTPUT)
    return

main()
```
